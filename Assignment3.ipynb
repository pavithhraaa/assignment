```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Sales': [150, 200, 250, 300, 350],
    'Profit': [50, 70, 90, 120, 150]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Sales'], color='blue', label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.legend()
plt.show()



```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Sales'], marker='o', color='blue', label='Sales')
plt.plot(df['Month'], df['Profit'], marker='o', color='green', label='Profit')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Profit')
plt.legend()
plt.show()
```
